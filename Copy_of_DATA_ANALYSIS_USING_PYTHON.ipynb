Calculate the percentage of
restaurants that serve each of the top
cuisines

```
def calculate_percentage(top_three_cuisines):
    total_restaurants = sum(top_three_cuisines.values())
    percentages = {}

    for cuisines, count in top_three_cuisines.items():
        percentage = (count / total_restaurants) * 100
        percentages[cuisines] = percentage
    return percentages
top_three_cuisines = {
    'North Indian'         :936,
    'North Indian, Chinese':511,
    'Chinese'              :354
}



result = calculate_percentage(top_three_cuisines)
for cuisines, percentage in result.items():
  print(f"{cuisines}: {percentage:.2f}%")
```

Task: Online Delivery TASK 4

Determine the percentage of restaurants
that offer online delivery.

```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('/content/Dataset.csv')
```

```
df
```

```
delivery_column = None
rating_column = None
search_keywords = ['delivery', 'Online_order', 'has_delivery']
rating_keywords = ['rating', 'rate', 'score']
for column in df.columns:
  if any(keyword in column.lower() for keyword in search_keywords):
    delivery_column = column
  if any(keyword in column.lower() for keyword in rating_keywords):
    rating_column = column
if delivery_column is None:
  raise ValueError("No column found with delivery information.")
if rating_column is None:
  raise ValueError("No column found with restaurant ratings.")
```

```
rating_mapping = {
    'Poor': 1,
    'Average': 2,
    'Good': 3,
    'Very Good': 4,
    'Excellent': 5
}
df[rating_column] = df[rating_column].map(rating_mapping)
```

```
df[delivery_column] = df[delivery_column].map({'Yes': 1, 'No': 0}).fillna(df[delivery_column])
```

```
avg_rating_with_delivery = df[df[delivery_column] == 1][rating_column].mean()
avg_rating_without_delivery = df[df[delivery_column] == 0][rating_column].mean()

print(f"Average rating for restaurants with online delivery: {avg_rating_with_delivery:.2f}")
print(f"Average rating for restaurants without online delivery: {avg_rating_without_delivery:.2f}")
```

LEVEL 2 TASK2 Task: Cuisine Combination

Identify the most common combinations of
cuisines in the dataset.

```
import pandas as pd
from itertools import combinations
from collections import Counter
```

```
df = pd.read_csv('/content/Dataset.csv')
```

```
df
```

```
def preprocess_cuisines(Cuisines):
  if pd.notnull(Cuisines):
    return str(Cuisines).strip().split(', ')
  else:
    return []
df['Cuisines'] = df['Cuisines'].apply(preprocess_cuisines)
```

```
all_combinations = df['Cuisines'].apply(lambda x: list(combinations(x, 2)))
combinations_counts = Counter([tuple(sorted(item)) for sublist in all_combinations for item in sublist])
```

```
most_common_combination = combinations_counts.most_common(1)[0]
print("The most combination of cuisines is:", most_common_combination)
```

```
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/content/Dataset.csv')
df['Rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')
df.dropna(subset=['Aggregate rating'], inplace=True)
expanded_df = df.explode('Cuisines')
average_ratings = expanded_df.groupby('Cuisines')['Aggregate rating'].mean().reset_index()

average_ratings.sort_values(by='Aggregate rating', ascending=False, inplace=True)


plt.figure(figsize=(10, 6))
plt.bar(average_ratings['Cuisines'], average_ratings['Aggregate rating'])
plt.xticks(rotation=45, ha='right')
plt.xlabel('Cuisine Combination')
plt.ylabel('Higher Rating')
plt.title('Higher Ratings of Cuisine combination')
plt.tight_layout()
plt.show()
```

LEVEL 2 TASK 4 RESTAURANT CHAINS

```
import pandas as pd
```

```
try:
    df= pd.read_csv('/content/Dataset.csv')
except FileNotFoundError:
    print("Error: '/content/Dataset.csv' file not found.")
    exit()
```

```
df
```

```

df['Restaurant Name'] = df['Restaurant Name'].str.strip()
df['Restaurant Name'] = df['Restaurant Name'].str.lower()
```

```
chain_restaurants = {}
for _, restaurant in df.iterrows():
    restaurant_name = restaurant['Restaurant Name']
    latitude = restaurant['Latitude']
    longitude = restaurant['Longitude']
    if restaurant_name in chain_restaurants:
        chain_restaurants[restaurant_name].append((latitude, longitude))
    else:
      chain_restaurants[restaurant_name] = [(latitude, longitude)]
```

```
print("Identified Restaurant Chains:")
for chain, locations in chain_restaurants.items():
    if len(locations) > 1:
      print(f"Chain: {chain}, Locations: {locations}")
```

```
import pandas as pd
import matplotlib.pyplot as plt


try:
    data = pd.read_csv('/content/Dataset.csv')
except FileNotFoundError:
    print("Error: '/Dataset (2).csv' file not found.")
    exit()

data['Restaurant Name'] = data['Restaurant Name'].str.strip()
data['Restaurant Name'] = data['Restaurant Name'].str.lower()

print(data.columns)
chain_data = data.groupby('Restaurant Name').agg({
    'Aggregate rating': 'mean',
    'Latitude': 'first',
    'Longitude': 'first',
    'Latitude': 'size'
}).reset_index()
chain_data.rename(columns={'Latitude': 'Popularity'}, inplace=True)
chain_data = chain_data.sort_values(by='Popularity', ascending=False)

print("Analyzed Ratings and Popularity of Restaurant Chains:")
print(chain_data[['Restaurant Name', 'Aggregate rating', 'Popularity']])

chain_data.to_csv('restaurant_chain_analysis.csv', index=False)

plt.figure(figsize=(10, 6))
plt.bar(chain_data['Restaurant Name'], chain_data['Popularity'])
plt.xlabel('Restaurant Chain')
plt.ylabel('Number of Locations (Popularity)')
plt.title('Popularity of Restaurant Chains')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```
